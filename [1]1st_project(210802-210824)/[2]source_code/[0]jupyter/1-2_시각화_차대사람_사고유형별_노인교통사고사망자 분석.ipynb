```python
## 주석 설명 
# 횡단중 : 횡단보도,무단횡단(인도와 차도, 모든 도로를 포함해서 걷고있을때)
# 차도통행중 : 건너려는것이아니라 도로(인도,차도)에 서 있을때(건너는의도가 없을때)
# 길 가장자리 구역 통행중 : 보도블럭이 없는곳, 차도와 인도사이의 선으로 처리해놓은 구역에서 일어난 사고, 
#                           보도블럭은 없지만 선으로 구분해놓은 곳 
# 보도 통행중 : 자동차가 인도를 침범했을때 일어난 사고 
# 기타 : 앞에서 언급한 내용을 제외한 모든 사건 (씽크홀,훼손도로로 일어난 사건 등)
```

```python
car_data_result.plot(kind='bar',color=('#ff2258','#4caf50','#01579b'),figsize=(10,10),rot=0,alpha=0.5)
                     
# car_data_result.title['사고유형별 발생건수']

plt.xlabel("사고유형",labelpad=20,fontsize=15)
plt.ylabel("발생건수",labelpad=20,fontsize=15)
plt.title("<사고유형별 노인 교통사고 사망 발생건수>",fontsize=20,pad=30,fontweight="bold")


plt.savefig('../[0]data/1-2-1_시각화_사고유형별_노인_교통사고_사망_발생건수(결과).png')
plt.show()
```
