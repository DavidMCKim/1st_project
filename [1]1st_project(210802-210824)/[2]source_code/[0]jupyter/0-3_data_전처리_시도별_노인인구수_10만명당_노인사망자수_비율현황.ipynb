## 지역별 65세이상 인구수 추출

```python
# 지역별로 각각 102개 행으로 나뉨
# data1.head(102) # : 전국 (계~100세이상)
# data1.tail(102) # : 제주특별자치도 (계~100세이상)

# 지역별로
# 전국 65세이상 인덱스 : 066 ~ 102
# 서울 65세이상 인덱스 : 168 ~ 204 => 66+102 ~ 102+102
# 부산 65세이상 인덱스 : 270 ~ 306 => 168+102 ~ 204+102
# 대구 65세이상 인덱스 : 372 ~ 408 => 270+102 ~ 306+102
# ...

# 위에서 보이는 규칙과 같이 for문 생성

# index : 각 지역의 시작점인 0,102,204,306,408,... 으로 설정

area_older_data = pd.DataFrame()

for index in range(0,len(kor_data_2019.index),102): # 0,102,204,306,
    new_kor_data = kor_data_2019.iloc[index+66:index+102,0:3] # 지역별 연령별 인구수 추출
    area_older_data = area_older_data.append(new_kor_data) 
```

```python
area_older_data
```

```python
area_older_data['총인구수 (명)'] = area_older_data['총인구수 (명)'].astype('int')
```

```python
# area_older_data 중 '행정구역(시군구)별'열의 값 중 중복없이 리스트 형태로 추출
areaList = area_older_data['행정구역(시군구)별'].drop_duplicates().tolist() # 시도별 지역 name list
areaList
```

```python
# 지역별로 65세~100세이상의 인구수를 모두 더해 하나의 df로 생성하여 area_older_df에 추가

index=0
areaOlderSumData = []

for area in areaList:
    areaOlderSumData.append(area_older_data.iloc[index:index+36,2:].sum()[0])
    area_older_df = pd.DataFrame(areaOlderSumData)
    
    index += 36
```

```python
area_older_df
```

## 지역별 지자체인구 수 추출

```python
# 지역별로 각각 102개 행으로 나뉨
# data1.head(102) # : 전국 (계~100세이상)
# data1.tail(102) # : 제주특별자치도 (계~100세이상)

# 지역별로
# 전국 모든 연령 인덱스 : 001 ~ 102
# 서울 모든 연령 인덱스 : 102 ~ 204 => 000+102 ~ 102+102
# 부산 모든 연령 인덱스 : 204 ~ 306 => 102+102 ~ 204+102
# 대구 모든 연령 인덱스 : 306 ~ 408 => 204+102 ~ 306+102
# ...

# 위에서 보이는 규칙과 같이 for문 생성

# index : 각 지역의 시작점인 0,102,204,306,408,... 으로 설정
# 각 지역마다 맨 처음 값은 계가 있으므로 시작 인덱스에 1을 더한 값부터 인덱싱

area_total_data = pd.DataFrame()
for index in range(0,len(kor_data_2019.index),102): # 0,102,204,306,
    new_kor_data1 = kor_data_2019.iloc[index:index+102,0:3] # 지역별 연령별 인구수 추출
    area_total_data = area_total_data.append(new_kor_data1) 
```

```python
area_total_data
```

```python
area_total_data['총인구수 (명)'] = area_total_data['총인구수 (명)'].astype('int')
```

```python
# 지역별로 0세~100세이상의 인구수를 모두 더해 하나의 df로 생성하여 area_total_df에 추가

index=0

areaTotalSumData = []

for area in areaList:
    areaTotalSumData.append(area_total_data.iloc[index+1:index+102,2:].sum()[0])
    area_total_df = pd.DataFrame(areaTotalSumData)

    index += 102
```

```python
area_total_df
```

```python
data = pd.DataFrame()
data['65세이상'] = area_older_df[0].values.tolist()
data['지자체인구'] = area_total_df[0].values.tolist()
```

```python
data.drop(index=0,axis=0,inplace=True)
```
