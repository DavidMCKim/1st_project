### 0-2.연령별 사상자수(사망자수+부상자수)당 사망자 비율

```python
nondie_per_die_result = die_data.merge(nondie_data,on='연령층별')
nondie_per_die_result
```

```python
nondie_per_die_result.rename(columns={'2019_x':'사망자수',                       
                       '2019_y':'부상자수'
                         }, inplace=True)


# result.rename({columns[0]='구분1',
#               {columns[1]='인원1',
#               {columns[2]='구분2',
#               {columns[3]='인원2'},inplace=True)
```

```python
nondie_per_die_result
```

```python
nondie_per_die_result["사상자수대비 사망자 비율"]=((nondie_per_die_result['사망자수']/(nondie_per_die_result["사망자수"]+nondie_per_die_result["부상자수"]))*100).round(2)
```

```python
nondie_per_die_result
```

```python
nondie_per_die_result.to_csv('../[0]data/0-1_전국_연령층별_교통사고_사상자(결과).csv')
```

```python
# 결과 : 
# 부상자수는 경제활동 인구인 20세~60세가 높게나왔음
# 하지만 사상자수 대비 사망자 비율은 65세이상이 가장 높게 나왔음
# 이 결과 노인(65세이상)층은 교통사고를 당해도 사망으로 이어지는 경우가 높음(추정:회복력 차이에서 기인할것이라 추정됨)
```
